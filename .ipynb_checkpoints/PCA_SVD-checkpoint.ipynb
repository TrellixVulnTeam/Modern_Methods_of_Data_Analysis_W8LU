### Task 1
In your data set, select a subset of 3-6 features related to the same aspect and explain your choice (may be the same subset that was used for k-means clustering)

Let us load dataset.
In order to discover which song is the hit and which is not the hidden
ranking factor has been calculated with the help of singular value decomposition. We will choose such features, that our clusters will interpret popularity of songs. Let it be instrumentalness, energy, loudness and danceability. Depending on these features we can slightly define popularity of the song (if it is instrumental, energetic, loud and danceable).

```python
import numpy as np
import pandas as pd

# decades = ['60s','70s','80s','90s','00s','10s']
decades = ['90s']
data_list = []
for decade in decades:
    file_name = r'dataset-of-'+decade+r'.csv'
    spotify_data = pd.read_csv(file_name)
    spotify_data['decade'] = decade
    data_list.append(spotify_data)
spotify_data_merged = pd.concat(data_list)
incols = ["instrumentalness", "energy", "loudness","danceability"] 
spotify_data_merged.reset_index(drop=True, inplace=True)
spotify_data_merged.info()
```

```python
spotify_data_merged.columns.tolist()
```

### Task 2
1. Standardize the selected subset

```python
# standartize data
df = spotify_data_merged[incols].copy()
# rescale loudness column  
df[['loudness']] -= df[['loudness']].min()
df[['loudness']] /= df[['loudness']].max()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

class Range:
    def fit_transform(self, data):
        return (data - data.mean(axis=0)) / (data.max(axis=0) - data.min(axis=0))
```

2. compute its data scatter and SVD

```python
from numpy.linalg import svd

def svd_contributions(data):
    ds = (data ** 2).sum()
    Z, Mu, C = svd(data, full_matrices=False)
    natural_contributions = Mu ** 2
    per_cent_contributions = natural_contributions / ds * 100
    
    return natural_contributions, per_cent_contributions, ds
```

```python
for normalizer in [StandardScaler(), Range(), MinMaxScaler()]:
    natural_contributions, per_cent_contributions, ds = svd_contributions(normalizer.fit_transform(df.values))
    print(normalizer.__class__.__name__)
    print('Data Scatter: {}'.format(ds))
    print('Natural contributions: {}'.format(natural_contributions))
    print('Per cent contributions: {}\n'.format(per_cent_contributions))
```

### Task 3
Compute and interpret a hidden ranking factor behind the selected features. The factor should be expressed in a 0-100 rank scale (as well as the features – ranking normalization).

```python
df_z = StandardScaler().fit_transform(df)
df_rnk = MinMaxScaler().fit_transform(df)
df_rng = Range().fit_transform(df)
```

```python
Z, Mu, C = svd(df_rnk, full_matrices=False)
alpha = 1 / (-C[0]).sum()
ranks = df_rnk @ (-C[0]) * alpha
```

```python
spotify_data_rnk = spotify_data_merged.copy()
spotify_data_rnk['hrf'] = ranks
output_features = ['track','artist','danceability','energy','loudness','instrumentalness','target','decade','hrf']
spotify_data_rnk[output_features].sort_values(by='hrf', ascending=False).head(10)
```

```python
spotify_data_rnk[output_features].sort_values(by='hrf', ascending=True).head(10)
```

```python
print('C[0]:', -C[0])
print('Lowest ranks:', np.sort(ranks)[:10])
print('Highest ranks:', np.sort(ranks)[::-1][:10])
```

### Task 4
Visualize the data using two first principal components at the standardization with two versions of normalization: 
- (a) range normalization 
- (b) z-scoring.

At these visualizations, use a distinct shape/color for points representing a pre-specified by your group of objects. 

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


labels = spotify_data_merged['target'].values
unique_labels = np.unique(labels)
print('Range normalization')
pca_df = PCA(n_components=2, svd_solver='full').fit_transform(df_rng)
plt.figure(figsize=(10, 10))
for label in unique_labels:
    label_rows = np.where(labels == label)
    label_pca_df = pca_df[label_rows]
    plt.scatter(label_pca_df[:, 0], label_pca_df[:, 1], label=label)
plt.grid()
plt.legend()
```

```python
print('Z-score normalization')
pca_df = PCA(n_components=2, svd_solver='full').fit_transform(df_z)
plt.figure(figsize=(10, 10))
for label in unique_labels:
    label_rows = np.where(labels == label)
    label_pca_df = pca_df[label_rows]
    plt.scatter(label_pca_df[:, 0], label_pca_df[:, 1], label=label)
plt.grid()
plt.legend()
plt.title("Sklearn PCA of Z-scoring Standardized Data")
```

```python
centered_df =  df_z - df_z.mean(axis=0)
print(type(centered_df))
N = centered_df.shape[0]
covariance_matrix = centered_df.T @ centered_df / N
eig_values, eig_vectors = np.linalg.eig(covariance_matrix)
max_eig_value = eig_values.max()
max_eig_vector = eig_vectors[:, eig_values.argmax()]
pc_first = centered_df @ max_eig_vector / np.sqrt(N * max_eig_value)
covariance_matrix = covariance_matrix - (max_eig_value * (max_eig_vector[:, None] @ max_eig_vector[None, :]))
eig_values, eig_vectors = np.linalg.eig(covariance_matrix)
max_eig_value = eig_values.max()
max_eig_vector = eig_vectors[:, eig_values.argmax()]
pc_second = centered_df @ max_eig_vector / np.sqrt(N * max_eig_value)
plt.figure(figsize=(10, 10))
for label in unique_labels:
    label_rows = np.where(labels == label)
    plt.scatter(pc_first[label_rows], -pc_second[label_rows], label=label)
plt.grid()
plt.legend()
plt.title("Conventional PCA of Z-Scored")
```

```python
Z_zscore, Mu_zscore, C = svd(df_z, full_matrices=False)
z_z0 = Z_zscore[:, 0]*np.sqrt(Mu_zscore[0])
z_z1 = -Z_zscore[:, 1]*np.sqrt(Mu_zscore[1])

plt.figure(figsize=(10, 10))
for label in unique_labels:
    label_rows = np.where(labels == label)
    label_z_z0 = z_z0[label_rows]
    label_z_z1 = z_z1[label_rows]
    plt.scatter(label_z_z0, label_z_z1, label=label)
plt.grid()
plt.title("PCA of Z-scoring Standardized Data")
```
